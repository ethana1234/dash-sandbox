```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

### Creating a line chart from list

```python
plt.plot(range(1, 10), [1,2,3,4,0,4,3,2,1])
```

### Line Chart with a Pandas dataframe

```python
cars = pd.read_csv('Data/mtcars.csv')
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = cars.mpg
```

```python
mpg.plot()
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot()
```

### Bar Chart from a list

```python
plt.bar(range(1, 10), [1,2,3,4,0,4,3,2,1])
```

### Bar Chart from a Pandas Dataframe

```python
mpg.plot(kind='bar')
```

```python
mpg.plot(kind='barh')
```

### Creating a pie chart from a list

```python
plt.pie([1,2,3,4,.5])
plt.show()
```

### Saving a plot

```python
plt.savefig('Data/pie_chart.png')
```

### Simple times series

```python
df = pd.read_csv('Data/Superstore-Sales.csv', index_col='Order Date', encoding='cp1252', parse_dates=True)
df.head(5)
```

```python
df2 = df.sample(n=100, random_state=25, axis=0)

plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Super Store Sales')

df2['Order Quantity'].plot()
```

### Eyeballing distributions with histograms

```python
cars.index = cars.car_names
mpg = cars.mpg
mpg.plot(kind='hist') # or plt.hist(mpg) and plt.show()
```

```python
sb.distplot(mpg)
```

### Scatter plots 

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
sb.regplot(data=cars, x='hp', y='mpg', scatter=True)
```

### Seaborn scatter plot matrix

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
plt.show()
```

### Boxplots

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sb.boxplot(data=cars, x='am', y='mpg', palette='hls')
```
